```python

x=arange(0.,5+0.001,0.01)
L=1
f=lambda x: L*exp(-L*x)
F=lambda x: -exp(-L*x)+1
Finv=lambda z: -log(1-z)/L

     
```

```python

         

plot(x,f(x),label='True distribution',lw=3)
plot(x,F(x),label='Cumulative function',lw=6)
legend(loc=0)
```

```python
N=10000
z=numpy.random.random_sample(N)
xsamp=Finv(z)
plot(x,f(x),lw=7)
hist(xsamp,50,normed =True)
```
