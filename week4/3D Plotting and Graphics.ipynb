```python
k=8.987*10**9#Nm^2/c^2
q=exp(9)
d1=2
dx=9/101.
dy=9/101.
x,y=meshgrid(arange(-4,4+dx,dx),arange(-4,4+dy,dy))
Vxy=(k*q)/(x**2*(y-d1)**2)**(1/2.)-(k*q)/(x**2*(y+d1)**2)**(1/2.)
fig=figure()
ax=fig.gca(projection='3d')
p=ax.plot_surface(x,y,Vxy,rstride=1,cstride=1,linewidth=0,cmap=cm.coolwarm)
```

```python
def dipolePotential(x,y,q,d):
    k=8.987*10**9#Nm^2/c^2
    Vxy=(k*q)/(x**2+(y-d)**2)**(1/2.)-(k*q)/(x**2+(y+d)**2)**(1/2.)
    return Vxy
```

```python
fig=figure()
ax=fig.gca(projection='3d')
p=ax.plot_surface(x,y,pointPotential(x,y,1,0.,0.),rstride=1,cstride=1,linewidth=0,cmap=cm.coolwarm)
```

```python
fig=figure()
ax=fig.gca(projection='3d')
p=ax.plot_surface(x,y,dipolePotential(x,y,exp(9),1),rstride=1,cstride=1,linewidth=0,cmap=cm.coolwarm)
```
